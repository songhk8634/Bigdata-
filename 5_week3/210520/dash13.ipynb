```python
import pandas as pd
import numpy as np
import dash
import plotly.express as px
import dash_core_components as dcc
import dash_html_components as html
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output



app = dash.Dash(__name__)

app.layout = html.Div(
    [
        html.Label('num1'),
        dcc.Input(
            id = 'text1-input',
            type = 'text'
        ),
        html.Label('num2'),
        dcc.Input(
            id = 'text2-input',
            type = 'text'
        ),
        html.Div(
            id = 'output'
        )

    ]
)
```

```python
@app.callback(
    Output('output', 'children'),
    [Input('text1-input','value'),
    Input('text2-input','value')]
)
def update_output(num1,num2):
    num1 = float(num1)
    num2 = float(num2)
    p = num1+num2
    m = num1-num2
    c = num1*num2
    d = num1/num2
    string1 = '{} + {} = {}입니다.\n'.format(num1,num2,p)
    string2 = '{} - {} = {}입니다.\n'.format(num1,num2,m)
    string3 = '{} * {} = {}입니다.\n'.format(num1,num2,c)
    string4 = '{} / {} = {}입니다.\n'.format(num1,num2,d)
    
    x = html.H2(string1+string2+string3+string4)
    return x
```

```python
app.run_server(host= '127.0.0.1',port = '7777')
```

```python

app = dash.Dash('__name__')
app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'my-dropdown',
            options = [{'label' : '{}단'.format(i), 'value' : i} for i in range(1,10)]
        ),
        dash_table.DataTable(
            id = 'my-table',
            page_size = 9
        )
    ]
)
```

```python
@app.callback(
    [Output('my-table','columns'),
    Output('my-table','data')],
    [Input('my-dropdown','value')]
)

def update_table(sets):
    new_df = df.loc[(df.a <=sets) & (df.a > sets-1),:]
    new_df = pd.pivot(new_df,index = 'val', columns = 'sets', values = 'result')
    new_df.columns = [str(i) for i in new_df.columns]
    columns = [{'name' : i, 'id': i} for i in new_df.columns]
    data = new_df.to_dict('records')
    return columns,data
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```
