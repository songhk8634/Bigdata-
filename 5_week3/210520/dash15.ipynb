```python
import pandas as pd
import numpy as np
import cx_Oracle
from sqlalchemy import create_engine

import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_table
from dash.dependencies import Input,Output,State
```

```python
conn = create_engine('oracle://c##big:123@localhost:1521/orcl')
```

```python
df = pd.read_sql_query('''select * from all_tab_cols where owner = 'C##BIG' ''', con= conn)
```

```python
app = dash.Dash('__name__')

app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'my-dropdown',
            options = [{'label' : i, 'value' : i} for i in df.table_name.unique()]
        ),
        html.Hr(),
        html.Br(),
        dash_table.DataTable(
            id = 'my-table'
        )
    ]
)
```

```python
@app.callback(
    Output('my-table','columns'),
    Output('my-table','data'),
    Input('my-dropdown','value')
)
def update_table(table_name):
    columns =  [{'name' : i.lower(), 'id' : i.lower()} for i in df.loc[df.table_name==table_name,'column_name']]
    stmt = 'select * from {}'.format(table_name)
    table_df = pd.read_sql_query(stmt,con=conn)
    data = table_df.to_dict('records')
    return columns,data
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```
