```python
import numpy as np
import pandas as pd
import plotly.express as px

import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_table
from dash.dependencies import Input, Output
```

```python
x = []
for i in range(1,10):
    x.extend([i]*9)
df = pd.DataFrame(x)
df.columns = ['sets']
df['val'] = (list(range(1,10))*9)
df['result'] = df.sets * df.val
```

```python
app = dash.Dash('__name__')
app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'my-dropdown',
            options = [{'label':'1-3단','value':3},
                      {'label':'4-6단','value':6},
                      {'label':'7-9단','value':9}]
        ),
        dash_table.DataTable(
            id = 'my-table',
            page_size = 9
        )
    ]
)
```

```python
@app.callback(
    [Output('my-table','columns'),
     Output('my-table','data')],
    [Input('my-dropdown','value')]
)
def update_table(sets):
    new_df = df.loc[(df.sets <= sets) & (df.sets > sets-3),:]
    new_df = pd.pivot(new_df,index='val',columns='sets',values='result')
    new_df.columns = [str(i)+'단' for i in new_df.columns]
    columns =[{'name':i, 'id':i} for i in  new_df.columns]
    data = new_df.to_dict('records')
    return columns, data


```

```python
app.run_server(host='192.168.0.7',port='7777')
```
