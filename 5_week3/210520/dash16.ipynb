# **원하는 Table과 원하는 Column**

```python
import pandas as pd
import numpy as np
import plotly.express as px

import cx_Oracle
from sqlalchemy import create_engine

import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_table
from dash.dependencies import Input,Output,State
```

```python
conn = create_engine('oracle://c##big:123@localhost:1521/orcl')

db_schema_qry = '''select * from all_tab_cols where owner = 'C##BIG' '''
db_schema = pd.read_sql_query(db_schema_qry, con=conn)
```

```python
app = dash.Dash('__name__')

app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'table-dropdown',
            options = [{'label' : i, 'value' : i} for i in db_schema.table_name.unique()]
        ),
        html.Hr(),
        dcc.Dropdown(
            id = 'column-dropdown'
        ),
        dash_table.DataTable(
            id = 'my-table'
        )
    ]
)
```

```python
@app.callback(
    Output('column-dropdown','options'),
    Input('table-dropdown','value')
)
def update_column (table_name):
    options =  [{'label' : i, 'value' : i} for i in db_schema.loc[db_schema.table_name==table_name,'column_name']]
    return options

@app.callback(
    Output('my-table','columns'),
    Output('my-table','data'),
    Input('table-dropdown','value'),
    Input('column-dropdown','value')
)
def update_table(table_name,column_name):
    columns =  [{'name' : column_name.lower(), 'id' : column_name.lower()}]
    # columns =  [{'name' : i.lower(), 'id' : i.lower()} for i in db_schema.loc[db_schema.table_name==table_name,'column_name']]
    stmt = 'select {} from {}'.format(column_name,table_name)
    table_df = pd.read_sql_query(stmt,con=conn)
    data = table_df.to_dict('records')
    return columns,data
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```
