```python
import pandas as pd               ## 데이터처리 
import numpy as np                ## 프로그래밍적으로 필요
import plotly.express as px       ## 시각화

import cx_Oracle                         ## 오라클 데이터베이스 접속
from sqlalchemy import create_engine     ## sql query 및 데이터베이스 접속

## dash : web programming 
import dash
import dash_html_components as html  ## html 용
import dash_core_components as dcc   ## html 컴포넌트 그리기
import dash_table                    ## table - grid 컴포넌트
from dash.dependencies import Input, Output, State  ## dash back-end용
```

```python
conn = create_engine('oracle://c##big:1234@localhost:1521/orcl')
db_schema_qry = '''select * from all_tab_cols where owner = 'C##BIG'  '''
db_schema = pd.read_sql_query(db_schema_qry,con=conn)
```

```python
app = dash.Dash(__name__)
app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'table-dropdown',
            options = [{'label':i,'value':i} for i in db_schema.table_name.unique()]
        ),
        html.Hr(),
        dcc.Dropdown(
            id = 'column-dropdown',
            multi = True
        ),
        html.Button(
            'submit',
            id = 'my-submit',
            n_clicks = 0
        ),
        dash_table.DataTable(
            id = 'my-table'
        )
    ]
)
```

```python
@app.callback(
    Output('column-dropdown','options'),
    Input('table-dropdown','value')
)
def update_column_dropdown(table_name):
    column_df = db_schema.loc[db_schema.table_name==table_name,'column_name']
    options = [{'label':i,'value':i} for i in column_df]
    return options

@app.callback(
    Output('my-table','columns'),
    Input('my-submit','n_clicks'),
    State('table-dropdown','value'),
    State('column-dropdown','value')
)
def update_table(number,table_name,column_name):
    columns = [{'name':i,'id':i} for i in column_name]
    return columns


```

```python
app.run_server(host='192.168.0.7',port='7777')
```
