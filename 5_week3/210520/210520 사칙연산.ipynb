```python
import pandas as pd
import numpy as np
import plotly.express as px

import dash
import dash_html_components as html
import dash_core_components as dcc
import dash_bootstrap_components as dbc
import dash_table
from dash.dependencies import Input, Output, State
```

```python
app = dash.Dash(__name__)

app.layout = html.Div(
    [
        html.Label('Num1'),
        dcc.Input(
            id = 'text1-input',
            type = 'text'
        ),
        html.Label('Num2'),
        dcc.Input(
            id = 'text2-input',
            type = 'text'
        ),
        html.Button(
            '계산',
            id = 'cal-button',
            n_clicks = 0,
            style = {'background-color':'DodgerBlue',
                    'color':'white'}
        ),
        html.Hr(),
        html.Div(
            [
                dash_table.DataTable(
                    id = 'my-table',
                    columns = [
                        {'name':"사용숫자",'id':"사용숫자"},
                        {'name':"연산",'id':"연산"},
                        {'name':"계산결과",'id':"계산결과"}
                    ],
                    style_cell={'minWidth': 95, 'maxWidth': 95, 'width': 9},
                    editable=True,
                    filter_action="native",
                    sort_action="native"
             )
        ]  
    )
    ]
)
```

```python

@app.callback(
    Output('my-table','data'),
    Input('cal-button','n_clicks'),
    State('text1-input','value'),
    State('text2-input','value')
)
def update_table(n,num1,num2):
    num1 = int(num1)
    num2 = int(num2)
    plus = num1+num2
    minus = num1 - num2
    multi = num1 * num2
    div = num1/num2
    data = [{"사용숫자":str(num1)+' ,'+str(num2),"연산":'덧셈',"계산결과":plus},
           {"사용숫자":str(num1)+' ,'+str(num2),"연산":'뺄샘',"계산결과":minus},
           {"사용숫자":str(num1)+' ,'+str(num2),"연산":'곱샘',"계산결과":multi},
           {"사용숫자":str(num1)+' ,'+str(num2),"연산":'나눗셈',"계산결과":div}]
    return data
```
