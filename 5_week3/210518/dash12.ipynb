# **Iris를 Table로 표현**

```python
import pandas as pd
import numpy as np
import plotly.express as px

import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_table
from dash.dependencies import Input, Output, State
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
pd.DataFrame(iris['data'], columns = iris['feature_names'])
```

```python
app = dash.Dash(__name__)
```

```python
app.layout = html.Div(
    [
        dash_table.DataTable(
            id = 'my-table',
            columns = [ {'name' : i, 'id' : i} for i in iris['feature_names']],
            data = pd.DataFrame(iris['data'], columns = iris['feature_names']).to_dict('records'),
            page_size = 6
        )
    ]
)
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```

# **Gapminder를 Table로 표현**

```python
import pandas as pd
import numpy as np
import plotly.express as px

import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_table
from dash.dependencies import Input, Output, State
```

```python
df = px.data.gapminder()
type(df)
```

```python
app = dash.Dash(__name__)
```

```python
app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'continent_dropdown',
            options = [
                {"label": i,"value": i } for i in df.continent.unique()
            ],
            value = df.continent.unique()[0]
        ),
        dcc.Dropdown(
            id = 'country_dropdown'
        ),
        dash_table.DataTable(
            id = 'gap-table',
            columns = [ {'name' : i, 'id' : i} for i in df.columns],           ## table 의 column
            data = df.to_dict('records'),                                       ## column name과 dictionary key와 쌍
            page_size = 10
        )
    ]
)
```

```python
@app.callback(
    Output('country_dropdown','options'),
    [Input('continent_dropdown','value')]
)

def update_country(continent):
    options = [
        {"label": i,"value": i } for i in df.loc[df.continent == continent, 'country'].unique()
    ]
    
    return options
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```

# **대륙, 나라 선택 후 Line Update (Table은 그대로)**

```python
import pandas as pd
import numpy as np
import plotly.express as px

import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_table
from dash.dependencies import Input, Output, State
```

```python
df = px.data.gapminder()
type(df)
app = dash.Dash(__name__)
```

```python
app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'continent_dropdown',
            options = [
                {"label": i,"value": i } for i in df.continent.unique()
            ],
            value = df.continent.unique()[0]
        ),
        dcc.Dropdown(
            id = 'country_dropdown'
        ),
        dash_table.DataTable(
            id = 'gap-table',
            columns = [ {'name' : i, 'id' : i} for i in df.columns],           ## table 의 column
            data = df.to_dict('records'),                                       ## column name과 dictionary key와 쌍
            page_size = 10
        ),
        dcc.Graph(
            id = 'country_line'
        )
    ]
)
```

```python
@app.callback(
    Output('country_dropdown','options'),
    [Input('continent_dropdown','value')]
)

def update_country(continent):
    options = [
        {"label": i,"value": i } for i in df.loc[df.continent == continent, 'country'].unique()
    ]
    
    return options


@app.callback(
    Output("country_line", "figure"), 
    [Input("country_dropdown", "value")]
)
def update_line_chart(country):
    fig = px.line(df[df.country==country], 
        x="year", y="pop", color='continent')
    return fig
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```

# **대륙, 나라 선택 후 Table, Line Update**

```python
# module import
import pandas as pd
import numpy as np
import plotly.express as px

import dash
import dash_core_components as dcc
import dash_html_components as html
import dash_table
from dash.dependencies import Input, Output, State
```

```python
df = px.data.gapminder()
app = dash.Dash(__name__)
```

```python
app.layout = html.Div(
    [
        dcc.Dropdown(
            id = 'continent_dropdown',
            options = [
                {"label": i,"value": i } for i in df.continent.unique()
            ],
            value = df.continent.unique()[0]
        ),
        dcc.Dropdown(
            id = 'country_dropdown'
        ),
        dash_table.DataTable(
            id = 'gap-table',
            columns = [ {'name' : i, 'id' : i} for i in df.columns],           ## table 의 column
            data = df.to_dict('records'),                                       ## column name과 dictionary key와 쌍
            page_size = 10
        ),
        dcc.Graph(
            id = 'country_line'
        )
    ]
)
```

```python
@app.callback(
    Output('country_dropdown','options'),
    [Input('continent_dropdown','value')]
)

def update_country(continent):
    options = [
        {"label": i,"value": i } for i in df.loc[df.continent == continent, 'country'].unique()
    ]
    
    return options


@app.callback(
    Output("country_line", "figure"), 
    [Input("country_dropdown", "value")]
)
def update_line_chart(country):
    fig = px.line(df[df.country==country], 
        x="year", y="pop")
    return fig


@app.callback(
    Output('gap-table','data'),
    [Input('country_dropdown','value')]
)
def update_table(country):
    dt = df.loc[df.country==country,:]
    data = dt.to_dict('records')
    return data
```

```python
app.run_server(host = '127.0.0.1', port = '7777')
```
