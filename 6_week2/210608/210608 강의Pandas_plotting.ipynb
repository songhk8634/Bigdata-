```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'], columns=iris.feature_names)
```

```python
iris_df.columns = ['sepalL','sepalW','petalL','petalW']
```

```python
iris_df['target'] = iris['target']
```

```python
iris_df['target2'] = np.where(iris_df.target == 0,iris.target_names[0],
        np.where(iris_df.target == 1, iris.target_names[1], iris.target_names[2]))
```

```python
iris_df.head(2)
```

#### iris 데이터셋 4개의 feature에 대하여 2개의 캔버스에 각각 sepal_length와 petal_length, sepal_width와 petal_width를 짝지어서 선 그래프로 plotting 해보세요

```python
x_axis = iris_df.index
```

```python
fig, ax = plt.subplots(1,2, sharey=True, figsize = (12,6))

ax[0].plot(x_axis, iris_df['sepalL'], label = iris_df.columns[0], c = 'r')
ax[0].plot(x_axis, iris_df['petalL'], label = iris_df.columns[2], c = 'b')
ax[1].plot(x_axis, iris_df['sepalW'], label = iris_df.columns[1], c = 'r')
ax[1].plot(x_axis, iris_df['petalW'], label = iris_df.columns[3], c = 'b')
ax[1].xaxis.set_ticks(np.arange(0.,150.1,30.))

ax[0].set_xlabel('index')
ax[1].set_xlabel('index')
ax[0].set_ylabel('legnth(cm)')
ax[1].set_ylabel('width(cm)')

ax[0].legend()
ax[1].legend(fontsize = 15)

ax[0].set_title('sepal_length - petal_length', fontsize = 15)
ax[1].set_title('sepal_width - petal_width')

plt.show()
```

#### pandas plot

```python
fig, ax = plt.subplots(1,2, sharey = True, figsize = (12,6))

iris_df[['sepalL','petalL']].plot(ax=ax[0], cmap = 'plasma')
iris_df[['sepalW','petalW']].plot(ax=ax[1], cmap = 'plasma', ls = '-.')

ax[0].set_xlabel('index')
ax[1].set_xlabel('index')

ax[0].set_ylabel('legnth(cm)')
ax[1].set_ylabel('width(cm)')

plt.show()
```
