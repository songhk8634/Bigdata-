# **Post 방식 크롤링**

목표 : 한국금거래소의 일별 금 가격 전체 데이터를 수집 (2008 ~ 현재) 하여 시각화하고, 시기별 특정 이벤트를 annotation하여 가격 변동과의 관계를 살펴본다.  
  
  
[<U>한국금거래소</U>](https://www.koreagoldx.co.kr/main/html.php?agencyCode=&htmid=goods/gold_list.html)

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
import requests
```

```python
chrome_loc = 'C:/chrome/chromedriver.exe'
driver = webdriver.Chrome(chrome_loc)
driver.get('https://www.koreagoldx.co.kr/main/html.php?agencyCode=&htmid=goods/gold_list.html')
```

```python
total_button = '#contents > div.sub_content > div > div.service_cont > div.clear > ul.graph_dt_buttons > li:nth-child(6) > label'
driver.find_element_by_css_selector(total_button).click()
time.sleep(3)
```

[header 및 payload]

```python
header = {'User-Agent' : 'Mozilla/5.0'}
```

```python
payload = {'srchDt' : 'ALL',
          'type' : 'Au',
          'dataDateStart' : '2008.03.11',
          'dataDateEnd' : '2021.05.06'}
```

[헤더와 페이로드 포함한 페이지 요청 및 reseponse]

```python
rr = requests.post('https://www.koreagoldx.co.kr/main/html.php?agencyCode=&htmid=goods/gold_list.html', headers = header, data = payload)
```

[응답 페이지 소스 파싱하기]

```python
soup_test = BeautifulSoup(rr.content, 'html.parser')
```
