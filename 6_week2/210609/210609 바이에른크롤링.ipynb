```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import time
import numpy as np
import pandas as pd
chrome_loc = 'C:/chrome/chromedriver.exe'
driver = webdriver.Chrome(chrome_loc)
driver.get('https://namu.wiki/w/FC%20%EB%B0%94%EC%9D%B4%EC%97%90%EB%A5%B8%20%EB%AE%8C%ED%97%A8#s-2.1')
```

```python
football_html = driver.page_source
football = BeautifulSoup(football_html, 'html.parser')
```

```python
mun = football.select('div > div > span > dl > dd > div > div > table > tbody > tr')
```

```python
squad = []
for i in range(2,28):
    country = [mun[i].select('a.wiki-link-internal')[0]['title']][0]
    number = mun[i].select('div.wiki-paragraph')[0].text
    position = mun[i].select('div.wiki-paragraph')[2].text
    korean = mun[i].select('div.wiki-paragraph')[3].text
    rome = mun[i].select('div.wiki-paragraph')[4].text
    birth = mun[i].select('div.wiki-paragraph')[5].text
    phy = mun[i].select('div.wiki-paragraph')[6].text
    cont_s = mun[i].select('div.wiki-paragraph')[7].text
    cont_f = mun[i].select('div.wiki-paragraph')[8].text
    etc = mun[i].select('div.wiki-paragraph')[9].text
    # for j in range(10):
        # print(mun[i].select('div.wiki-paragraph')[j].text)
    #    number = mun[i].select('div.wiki-paragraph')[j].text
   
    squad.append([country, number, position, korean, rome, birth, phy, cont_s, cont_f, etc])
```

```python
squad = pd.DataFrame(squad)
squad.columns = ['국적', '등번호', '포지션', '한글', '로마자', '생년월일', '신체조건', '계약년도', '계약만료', '비고']
squad
```
