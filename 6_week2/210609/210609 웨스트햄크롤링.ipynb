```python
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import time
import numpy as np
import pandas as pd
chrome_loc = 'C:/chrome/chromedriver.exe'
driver = webdriver.Chrome(chrome_loc)
driver.get('https://en.wikipedia.org/wiki/West_Ham_United_F.C.#First-team_squad')
```

```python
west_html = driver.page_source
west = BeautifulSoup(west_html, 'html.parser')
```

```python
westham = []
for i in range(16):
    number = west.select('table.football-squad')[0].select('tr > td')[i*4].text[:-1]
    
    position = west.select('table.football-squad')[0].select('tr > td')[i*4 +1].select('a')[0].text
    country = west.select('table.football-squad')[0].select('tr > td')[i*4 +2].select('a')[0].text
    name = west.select('table.football-squad')[0].select('tr > td')[i*4 +3].select('a')[0].text
    
    westham.append([number, position, country, name])
    
for j in range(16):
    number = west.select('table.football-squad')[1].select('tr > td')[j*4].text[:-1]
    
    position = west.select('table.football-squad')[1].select('tr > td')[j*4 +1].select('a')[0].text
    country = west.select('table.football-squad')[1].select('tr > td')[j*4 +2].select('a')[0].text
    name = west.select('table.football-squad')[1].select('tr > td')[j*4 +3].select('a')[0].text
    
    westham.append([number, position, country, name])
```

```python
westham = pd.DataFrame(westham)
westham.columns = ['등번호', '포지션', '국적', '이름']
```
