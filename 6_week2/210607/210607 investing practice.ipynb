```python
from bs4 import BeautifulSoup
from selenium import webdriver
chrome_loc = 'C:/chrome/chromedriver.exe'
driver = webdriver.Chrome(chrome_loc)
url_target = 'https://kr.investing.com/equities/south-korea'
driver.get(url_target)
```

```python
invest_html = driver.page_source
soup = BeautifulSoup(invest_html, 'html.parser')
title2 = soup.select('td')   ## 나머지 뽑기위함
```

```python
[j.text for j in title2][1:8]
```

```python
name = []
now = []
high = []
low = []
delta = []
delta_p = []
count = []
for i in range(50):
    name.append([j.text for j in title2][1 + i*10])
    now.append([j.text for j in title2][2 + i*10])
    high.append([j.text for j in title2][3 + i*10])
    low.append([j.text for j in title2][4 + i*10])
    delta.append([j.text for j in title2][5 + i*10])
    delta_p.append([j.text for j in title2][6 + i*10])
    count.append([j.text for j in title2][7 + i*10])
```

```python
import pandas as pd
investing = pd.DataFrame()
investing['종목'] = name
investing['현재가'] = now
investing['고가'] = high
investing['저가'] = low
investing['변동'] = delta
investing['변동 %'] = delta_p
investing['거래량'] = count
investing.index = range(1,51)
```
