```python
import numpy as np
import pandas as pd
import warnings
warnings.simplefilter('ignore')

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input,Output,State
```

```python
X = load_iris()['data']
y = load_iris()['target']
```

```python
model = RandomForestClassifier()
model.fit(X,y)
```

```python
df = pd.DataFrame(load_iris()['data'])
df.columns = load_iris()['feature_names']
df['target'] = load_iris()['target']
species = np.where(df.target == 0, load_iris()['target_names'][0],
                  np.where(df.target == 1, load_iris()['target_names'][1], load_iris()['target_names'][2]))
```

```python
df['species'] = species
```

```python
df
```

```python
X = df.iloc[:, :-2]
y = df.iloc[:,-2]
model = RandomForestClassifier()
model.fit(X,y)
```

```python
sl_input = dcc.Input( 
    id = 'sl-input',
    type = 'text'
)
sw_input = dcc.Input(
    id = 'sw-input',
    type ='text'
)
pl_input = dcc.Input(
    id = 'pl-input',
    type = 'text'
)
pw_input = dcc.Input(
    id = 'pw-input',
    type = 'text'
)
analyze_button = html.Button(
    '분석',
    id = 'analyze-button',
    n_clicks = 0
)


app = dash.Dash(__name__)
app.layout = html.Div(
    [
        sl_input,
        sw_input,
        pl_input,
        pw_input,
        analyze_button,
        html.Div(
            id = 'html-output'
        )
    ]
)
```

```python
@app.callback(
    Output('html-output','children'),
    Input('analyze-button','n_clicks'),
    State('sl-input','value'),
    State('sw-input','value'),
    State('pl-input','value'),
    State('pw-input','value')
)
def iris_analyze(n,sl,sw,pl,pw):
    x = [[sl,sw,pl,pw]]
    result = model.predict(x)[0]
    if result ==0:
        species = 'Setosa'
    elif result ==1:
        species = 'Versicolor'
    else :
        species = 'Virginica'
    return html.H1('예측된 붓꽃의 품종은 {} 입니다.'.format(species))
```
