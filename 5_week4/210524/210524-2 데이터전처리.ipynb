## Pivot Tables

```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.isna().sum()
```

## Pivot Tables by Hand

```python
titanic.groupby('sex').mean()

titanic.groupby('sex')['survived'].mean()    ## Series 데이터

titanic.groupby('sex')[['survived']].mean()  ## DataFrame

titanic.groupby(['sex', 'class'])['survived'].aggregate('mean')     ## Series 데이터

titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()    ## DataFrame
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate('mean').reset_index()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].aggregate('mean').unstack()
```
