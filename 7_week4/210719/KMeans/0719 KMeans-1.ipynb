# **모듈 및 랜덤 데이터 생성**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, random_state = 100)
```

```python
data
```

```python
data[0][:,0]
```

```python
sns.scatterplot(x = data[0][:,0], y = data[0][:, 1])
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 4)
model.fit(data[0])
model.labels_
```

```python
data[1]
```

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = data[1], palette = 'Set2')
```

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = model.labels_, palette = 'Set2')
```

## **다른 K값으로 클러스터링 해보기**

```python
model = KMeans(n_clusters = 3)
model.fit(data[0])
plt.figure(figsize = (8, 6))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = model.labels_, palette = 'Set2')
```

```python
model = KMeans(n_clusters = 5)
model.fit(data[0])
plt.figure(figsize = (8, 6))
sns.scatterplot(x = data[0][:, 0], y = data[0][:, 1], hue = model.labels_, palette = 'Set2')
```

## **최적의 K값 찾기 : Elbow method**

```python
model.inertia_
```

```python
distance = []
for i in range(2, 11):
    model = KMeans(n_clusters = i)
    model.fit(data[0])
    distance.append(model.inertia_)
```

```python
distance
```

```python
list(range(2, 11))
```

```python
sns.lineplot(x = list(range(2, 11)), y = distance, marker = 'o')
```
