# **카테고리 변수 처리**

```python
col_list = []
for i in data.columns:
    if data[i].dtype == 'O':
        col_list.append(i)
```

```python
col_list
```

```python
for i in col_list:
    print(i, data[i].nunique())
```

```python
col_list = col_list[1:]
```

```python
data = pd.get_dummies(data, columns = col_list, drop_first = True)
```

```python
data.head(3)
```

# **Missing Value처리**

```python
data.isna().sum() / len(data)
```

```python
data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].mean())
```

```python
data.isna().sum()
```

## **Scaling**

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
# MinMax Scaler
minmax = MinMaxScaler()
```

```python
data.drop('customerID', axis = 1, inplace = True)
```

```python
minmax.fit(data)
scaled_data = minmax.transform(data)
```

```python
type(scaled_data)
```

```python
pd.DataFrame(scaled_data, columns = data.columns)
```

```python
# Robust Scaler
rob = RobustScaler()
rob.fit(data.drop('Churn_Yes', axis = 1))
scaled_rob = rob.transform(data.drop('Churn_Yes', axis = 1))
```

```python
type(scaled_rob)
```

```python
pd.DataFrame(scaled_rob, columns = data.drop('Churn_Yes', axis = 1).columns)
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X = scaled_rob
scaled_data = pd.DataFrame(scaled_data, columns = data.columns)
y = scaled_data['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

## **Modeling**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
pd.DataFrame({'actual_value' : y_test, 'pred_value' : pred})
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

### **최적의 K값 구하기**

```python
error_list = []
```

```python
for i in range(1, 101):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_list.append(accuracy_score(y_test, pred))
```

```python
plt.figure(figsize = (18, 10))
sns.lineplot(x = range(1, 101), y = error_list, 
             marker = 'o', markersize = 8, markerfacecolor = 'r')
```

```python
max(error_list)
```

```python
error_list.index(max(error_list))
```

```python
knn = KNeighborsClassifier(n_neighbors = 33)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print(accuracy_score(y_test, pred))
print(confusion_matrix(y_test, pred))
```
