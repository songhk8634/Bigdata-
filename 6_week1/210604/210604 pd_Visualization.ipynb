# **연습 2 & 3 데이터프레임 생성**
위와 같이 머신러닝을 위한 데이터는 numpy array로 준비하지만, 데이터 분석을 위해서는 불편함이 있다.

2. 따라서 분석에 용이하도록 데이터프레임으로 데이터를 생성한다.(4개 변수와 1개 target)

3. target2 columns을 추가하여 품종 이름 추가(0,1,2 -->setosa,...)

```python
from sklearn import datasets
iris = datasets.load_iris()
iris.keys()
```

```python
df_iris = pd.concat([pd.DataFrame(iris.data), pd.DataFrame(iris.target)], axis = 1,ignore_index=True)
```

```python
df_iris
```

```python
iris.feature_names.append('target')
```

```python
df_iris.columns = iris.feature_names
```

```python
df_iris['target2'] = np.where(df_iris.target == 0, iris.target_names[0],
                             np.where(df_iris.target == 1, iris.target_names[1],
                                     iris.target_names[2]))
```

```python
df_iris
```

# **pie chart**

```python
np.random.seed(10)
a = np.random.choice(list(df_iris.index), size = 30, replace = False)
a
```

```python
len(a)
```

```python
iris_df2 = df_iris.copy()
```

```python
iris_df2['target3'] = iris_df2['target2']
```

```python
iris_df2['target3'][a] = 'another'
```

```python
iris_df2['target3'].value_counts()
```

```python
labels = iris_df2['target3'].unique().tolist()
```

```python
labels
```

```python
sizes = iris_df2.target3.value_counts().tolist()
```

```python
plt.pie(sizes, labels = iris_df2['target3'].value_counts().index)
```

```python
plt.pie(sizes, labels = iris_df2['target3'].value_counts().index,
       shadow = True, explode = (0,0.2,0,0.2), counterclock = True,
       autopct = '%.1f')
plt.title('iris types', fontsize = 15)
plt.show()
```
